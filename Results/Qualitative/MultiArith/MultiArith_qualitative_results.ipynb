# Checking the matching digit and operator % for the baseline model

```python
baseline_preds_pass1 = baseline_preds[['lEquations',	'answer',	'question',	'context',	'equation0'	]]
baseline_preds_pass1.shape
```

```python
baseline_preds_pass1.tail()
```

```python
baseline_preds_pass1 = baseline_preds_pass1.drop(baseline_preds_pass1.index[-2:])  # Drop the last two rows
baseline_preds_pass1.shape
```

```python
# Function to extract digits from a string
def extract_digits(equation):
    # Extract digits using regular expression
    digits = re.findall(r'\d+\.\d*|\.\d+|\d+', equation)
    # Convert digits to integers
    digits = [int(float(digit)) for digit in digits]
    return digits

# Function to extract the operator from a string
def extract_operator(s):
    operators = set(['+', '-', '*', '/'])  # You can add more operators if needed
    for char in s:
        if char in operators:
            return char
    return None
```

```python
# Convert columns to strings
baseline_preds_pass1['lEquations'] = baseline_preds_pass1['lEquations'].astype(str)
baseline_preds_pass1['equation0'] = baseline_preds_pass1['equation0'].astype(str)

# Apply the functions row-wise to extract digits and operators
baseline_preds_pass1['digits1'] = baseline_preds_pass1['lEquations'].apply(extract_digits)
baseline_preds_pass1['operator1'] = baseline_preds_pass1['lEquations'].apply(extract_operator)
baseline_preds_pass1['digits2'] = baseline_preds_pass1['equation0'].apply(extract_digits)
baseline_preds_pass1['operator2'] = baseline_preds_pass1['equation0'].apply(extract_operator)

baseline_preds_pass1['digits1_len'] = baseline_preds_pass1['digits1'].apply(len)
baseline_preds_pass1['digits2_len'] = baseline_preds_pass1['digits2'].apply(len)

baseline_preds_pass1['operator1_len'] = baseline_preds_pass1['operator1'].apply(lambda x: len(x) if x is not None else 0)
baseline_preds_pass1['operator2_len'] = baseline_preds_pass1['operator2'].apply(lambda x: len(x) if x is not None else 0)
```

```python
# Compare the digits and operators
baseline_preds_pass1['matching_digits'] = baseline_preds_pass1.apply(lambda row: len(set(row['digits1']) & set(row['digits2'])), axis=1)
baseline_preds_pass1['matching_operators'] = baseline_preds_pass1.apply(lambda row: 1 if row['operator1'] == row['operator2'] else 0, axis=1)
```

```python
# Calculate the percentage of matching digits for each row
baseline_preds_pass1['matching_digit_percentage'] = baseline_preds_pass1.apply(
    lambda row: (row['matching_digits'] / max(row['digits1_len'], row['digits2_len'])) * 100
    if row['digits1_len'] is not None and row['digits2_len'] is not None else None,
    axis=1
)

# Calculate the percentage of matching digits for each row
baseline_preds_pass1['matching_operator_percentage'] = baseline_preds_pass1.apply(
    lambda row: (row['matching_operators'] / max(row['operator1_len'], row['operator2_len'])) * 100
    if row['operator1_len'] is not None and row['operator2_len'] is not None else None,
    axis=1
)
```

```python
baseline_preds_pass1.head(3)
```

```python
matching_digit_percentage_mean = baseline_preds_pass1['matching_digit_percentage'].mean()
print(f"Mean Matching digit Percentage: {matching_digit_percentage_mean}%")

matching_operator_percentage_mean = baseline_preds_pass1['matching_operator_percentage'].mean()
print(f"Mean Matching operator Percentage: {matching_operator_percentage_mean}%")
```
